```python
import ollama

ollama.list()
# ollama.show("llama3.1")
```

```python
ollama.chat(
    model="llama3.1", 
    messages=[{"role": "user", "content": "Why is the sky blue?"}]
)
```

```python
response = ollama.generate(
    model="llama3.1",
    prompt="""You're a clickbait detector. Detect if the given text is clickbait or non-clickbait. 
                Also, think step-by-step and extract the words that influenced the text to become clickbait or non-clickbait.
                Provide your response in the following JSON format:
                {
                    "text": <the input text>,
                    "verdict": <clickbait or non-clickbait>,
                    "words": <words that influenced the verdict>
                }
                Don't return any other text except the output format.
                The input text is given as follows:
                "The top 10 best selling comic books of all time [Video]"
                """,
)

print(response["response"])
```
